#### Overview

Central Limit Theorem considers sampling distribution in a manner similar to the law of large numbers. In this
case, instead of approaching true mean & zero variance, we standardize the sampling distribution to reshape it
to zero mean & unit variance.

It turns out that as long as the mean & variance of the original distribution are finite, the standardized
distribution built on its samples will converge to a Normal Distribution.

#### de Moivre â€“ Laplace CLT

A special of central limit theorem is the convergence of Binomial to Normal
distribution. Binomial distribution can be thought of as sampling distribution
with a number of identical independently distributed boolean random variables.

If we start with a Binomial Distribution Symmetric around its mean &
standardize it, we can easily show that for a large value of $n$, the
dominating terms of the distribution are identical to the standard Normal using
a series of two approximations.

Symmetric Binomial Distribution, Sterling's approximation, & Taylor's series for natual logarithm are defined as follows,

$$
    P(k) = \frac{1}{2^n} \frac{n!}{k!(n-k)!},
    \mu = n/2,
    \sigma^2 = n/4
$$

$$
    \ln(n!) = \sum_{i=1}^{n} \ln(i)
    \approx \int^n_1 \ln(x) dx
    \approx n \cdot \ln(n) - n
    \Rightarrow n! \approx n^n e^{-n}
$$

$$
    \ln(1+x) = \sum_{i=1}^{\infty} {(-1)}^{i-1}\frac{x^i}{i}
    \approx x - \frac{x^2}{2}, 0 < x \ll 1
$$

\
Applying Stirling's Approximation to symmetric Binomial definition, we get the
following,

$$
    \approx \frac{1}{2^n} \frac{n^n e^{-n}}{k^k e^{-k} (n-k)^{n-k} e^{-(n-k)}}
$$

$$
    \Rightarrow \frac{1}{2^n} \frac{n^n}{k^k(n-k)^{n-k}}
$$

$$
    \Rightarrow
    \left(\frac{n/2}{k}\right)^k
    \left(\frac{n/2}{n-k}\right)^{n-k}
$$

$$
    \Rightarrow
    \exp\left\{
    k \cdot \ln\left(\frac{n/2}{k}\right) +
    (n-k) \cdot \ln\left(\frac{n/2}{n-k}\right)
    \right\}
$$

Let $x = \frac{k - n/2}{\sqrt{n}/2}$ and approximate natural logarithm using Taylor's series,

$$
    \Rightarrow
    \exp\left\{\frac{1}{2}\left(
    -(n + x\sqrt{n}) \cdot \ln\left(1+\frac{x}{\sqrt{n}}\right) +
    -(n - x\sqrt{n}) \cdot \ln\left(1-\frac{x}{\sqrt{n}}\right)
    \right)\right\}
$$

$$
    \approx
    \exp\left\{\frac{1}{2}\left(
    -(n + x\sqrt{n}) \cdot \left(\frac{x}{\sqrt{n}} - \frac{x^2}{2n}\right) +
    -(n - x\sqrt{n}) \cdot \left(-\frac{x}{\sqrt{n}} - \frac{x^2}{2n}\right)
    \right)\right\}
$$

$$
    \Rightarrow e^{-\frac{x}{2}}
$$

$$
    \boxed{
        \therefore P(k) \approx e^{-\frac{x}{2}}
    }
$$

Note that this function is defined by the following
core property,

$$
    \boxed{
        \frac{d}{d\alpha}f(\alpha)= -\alpha \cdot f(\alpha)
    }
$$

If we have non-zero mean & non-unit variance, the following generalized
version holds,

$$
    \boxed{
        \frac{d}{d\alpha}f(\alpha)= -\frac{\alpha-\mu}{\sigma^2} \cdot
        f(\alpha)
    }
$$

We can show that the same holds for a Binomial distribution as follows,

$$
    P(k+1) - P(k) = \frac{n!}{(k+1)!(n-(k+1))!} - \frac{n!}{k!(n-k)!}
$$

$$
    \Rightarrow P(k)\left(\frac{n-k}{k+1} - 1\right)
    \Rightarrow P(k)\frac{n-2k-1}{k+1}
$$

$$
    \Rightarrow P(k)\frac{n-2(n/2 + x\sqrt{n}/2)-1}{n/2 + x\sqrt{n}/2 +1}
    \Rightarrow -P(k)\frac{x\sqrt{n} + 1}{n/2 + x\sqrt{n}/2 + 1}
    \approx -\frac{x}{\sqrt{n}/2}P(k)\frac{xn/2}{xn/2}
$$

$$
    \boxed{
        \therefore P(k+1)-P(k) \rightarrow -\frac{k-n/2}{n/4}P(k)
    }
$$


#### Lindeberg-L&eacute;vy CLT

The general CLT focuses on Samples taken from any probability distribution with
finite mean & variance. By analysing characterstic function or
moment-generating function, it can be seen that the structure approaches the
Normal Distribution.

Note that the Characterstic function of a distribution is analogous to Fourier
Transform & MGF is a variant of the same which abstracts away the process of
computing $n^{th}$ moment of the distribution.

Assume ${X_1,X_2,..X_n}$ are iid random variables with finite mean $\mu$ and
variance $\sigma^2$,

$$
  Z_n
  = \sum_{i=1}^{n}\frac{X_i - \mu}{\sqrt{n}\sigma}
  = \sum_{i=1}^{n}\frac{Y_i}{\sqrt{n}}
  ,\,\,
  E[Y_i] = 0, V[Y_i] = 1
$$

Focusing on the characterstic function of $Z_n$,

$$
  \phi_{Z_n}(t)
  = \phi_{\sum_{i=1}^{n}Y_i/\sqrt{n}}(t) = \phi_{Y_1}(t/\sqrt{n})^n
$$

$$
  \Rightarrow
  \,\,\phi_{Z_n}(t) =
  (1 + E[Y_1]\cdot it/\sqrt{n} - V[Y_1]\cdot t^2/2n + \epsilon)^n
$$

$$
  \boxed{
    \phi_{Z_n}(t) \approx e^{\frac{-t^2}{2}}
  }
$$

Focusing on the characterstic function of standard normal
distribution,

$$
  \phi_{N(0,1)}
  = \int \frac{1}{\sqrt{2\pi}} e^{\frac{-x^2}{2}} e^{itx} dx
  = \frac{1}{\sqrt{2\pi}} \int e^{itx  - \frac{x^2}{2}} dx
  = \frac{1}{\sqrt{2\pi}} \int e^{-\frac{1}{2} (x^2 - 2itx + (it)^2)} dx
$$

$$
  \Rightarrow \phi_{N(0,1)}
  = e^{\frac{-t^2}{2}} \frac{1}{\sqrt{2\pi}} \int e^{-\frac{1}{2} (x - it)^2}
  dx
$$

$$
  \boxed{
    \phi_{N(0,1)} = e^{\frac{-t^2}{2}}
  }
$$

$$
  \boxed{
    \therefore Z_n \rightarrow N(0, 1)
  }
$$
